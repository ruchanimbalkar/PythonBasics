#Write a Program that accepts user input of a positive angle and determines whether the angle is a right angle or or an acute angle or an obtuse angle etc.
**Theory**
Angles.jpg

#Pseudocode
Precondition: User enters positive integer values as angles.
Postcondition: We display the kind of angle on console.
1. Ask user for input ( a postive angle degree) "Please enter the size of your positive angle (0 < angle < 360)to determine its type: "
2. If input is not valid ( alphanumeric input or negative angle or 0)
    1. print "Invalid input."
3. Else (means valid input)
    1. Convert input to integer format
    2. If angle greater than smaller than 90 degree
         1. print "Acute angle".
    3. Else if check angle equals 90 degree
        1. print "Right angle."
    4. Else if check angle is greater than 90 degree but less than 180 degree
        1. print "Obtuse angle."
    5. Else if check angle equals 180 degree
        1. print "Straight angle."
    6. Else if check angle greater than 180 degree but less than 360 degree
        1. print "Reflex angle."
    7. Else (means angle greater than or equal to 360 degree )
        1. print "Out of scope."

```python
angle =  input("Please enter the size of your positive angle (0 < angle < 360)to determine its type: ")
if angle.isdecimal()==False or angle.isdecimal()=="0":
  print("Invalid input")
else:
  angle=int(angle)
  if angle < 90 :
    print("Acute angle")
  elif angle == 90:
    print("Right angle")
  elif angle > 90 and angle <180:
    print ("Obtuse angle")
  elif angle == 180 :
    print("Staright angle")
  elif angle > 180 and angle < 360 :
    print("Reflex angle")
  else: #angle >= 360
    print("Out of scope")
```
